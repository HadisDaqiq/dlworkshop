## Build the perceptron
To build and train a perceptron we have to perform three steps:
- Calculate the perceptron's output $\hat{y} = \left(\sum_i w_i X_i \geq 0\right)$
- Determine the update for the weights using the error and the learning rate: $\partial w_i = \eta (y-\hat{y}) X_i$
- Calculate new weights as: $w_i = w_i + \partial w_i$

```python
# initializes weights
w = np.random.rand(len(features))
print(w)
alpha = 1e-3
unit_step = lambda x: (x < 0)
```

```python
def calculate_perceptron_output(w, x):
    # Calculate the perceptrons output using 
    # np.dot(w, x) to calculate the sum and
    # thresholding the output:
    # solution: (np.dot(w, x) >= 0).astype(float)
    return (np.dot(w, x) >= 0).astype(float)
```

Let us change to pyTorch:

```python
from torch import nn
import torch
```

```python
class Perceptron(nn.Module):
    
    def __init__(self, number_of_inputs):
        super().__init__()
        self.linear = nn.Linear(number_of_inputs, 1, bias=False)
        
    def forward(self, x):
        return self.linear(x)

perceptron = Perceptron(len(features))
for name, param in perceptron.named_parameters():
    print(name, param)
```

```python

w = torch.from_numpy(np.random.rand(len(features)))
print(w)
x_ttrain = torch.from_numpy(x_train)
y_ttrain = torch.from_numpy(y_train.values)
for x, y in zip(x_ttrain, y_ttrain):
    print(w, x, y)
    
    result = torch.dot(w, x)
    error = y - unit_step(result)
    
    w += alpha * error * x
```

```python
results=[]
expected=[]
for x, y in zip(x_ttrain, y_ttrain):
    result = torch.dot(w, x)
    expected.append(y)
    results.append(result)
results = torch.stack(results)
expected = torch.stack(expected)
print(w)
print((unit_step(results) == expected.byte()).float().mean())
```
