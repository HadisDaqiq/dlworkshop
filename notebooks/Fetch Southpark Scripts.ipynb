```python
import requests_html
import pandas as pd
from requests_html import HTMLSession
from os.path import expanduser
home = expanduser("~")

session = HTMLSession()

class SouthParkEndpoint(object):
    _domain = 'https://southpark.fandom.com'

    scripts_page = _domain + '/wiki/Portal:Scripts'
    
    @staticmethod
    def get_seasons():
        r = session.get(SouthParkEndpoint.scripts_page)
        items = r.html.find('.wikia-gallery-item')
        season_links = []
        for item in items:
            season_links.append(item.find('a', first=True).attrs['href'])
        return season_links
        # print(items)
        
    @staticmethod
    def get_episodes(season_link):
        print(SouthParkEndpoint._domain + season_link)
        r = session.get(SouthParkEndpoint._domain + season_link)
        items = r.html.find('.wikia-gallery-item')
        episode_links = []
        for item in items:
            episode_links.append(item.find('a', first=True).attrs['href'])
        return episode_links
    
    @staticmethod
    def get_episode_text(episode_link, season, episode):
        r = session.get(SouthParkEndpoint._domain + episode_link)
        trs = r.html.find('table')[1].find('tr')
        content = []
        title = trs.pop(0).find('td')[0].full_text.strip()
        current_setting = ''
        for tr in trs:
            tds = tr.find('td')
            if len(tds) == 1: continue
            if not tds[0].full_text.strip():
                if tds[1].full_text.strip():
                    current_setting = tds[1].full_text.strip()
            else:
                ent = {}
                ent['character'] = tds[0].full_text.strip()
                full_text = tds[1].full_text.strip()
                spoken = ''
                actions = ''
                bracket_idx = full_text.find('[')
                while bracket_idx != -1:
                    spoken += full_text[:bracket_idx]
                    closing_idx = full_text.find(']')
                    if closing_idx == -1:
                        actions += full_text[bracket_idx:] + ';'
                        full_text = ''
                    else:
                        actions += full_text[bracket_idx:closing_idx+1] + ';'
                        full_text = full_text[closing_idx+1:]
                    bracket_idx = full_text.find('[')
                spoken += full_text
                ent['setting'] = current_setting
                ent['season'] = season
                ent['episode'] = episode
                ent['spoken'] = spoken
                ent['actions'] = actions
                content.append(ent)
        return content
        
        
        
        
all_data = []
seasons = SouthParkEndpoint.get_seasons()
for i, season_link in enumerate(seasons):
    print(f'Working on Season {i+1}')
    episodes = SouthParkEndpoint.get_episodes(season_link)
    for j, episode_link in enumerate(episodes):
        print(f'Working on Episode {j+1}')
        all_data.extend(SouthParkEndpoint.get_episode_text(episode_link, i+1, j+1))
pd.DataFrame(all_data).to_csv(home+'/data/workshop_data/southpark_full.csv', index=False, sep='\t')


```
